The dataset contains transactions made by credit cards ,This dataset presents transactions that occurred in two days, where we have 492 frauds out of 284,807 transactions. The dataset is highly unbalanced, the positive class (frauds) account for 0.172% of all transactions.

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('/Users/shero/Downloads/archive/creditcard.csv')
```

```python
#legit transactions & fraudulent transactions
df["Class"].value_counts()
```

```python
# 0 --> Normal Transaction
# 1 --> fraudulent transaction
legit = df[df.Class == 0]
fraud = df[df.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
fraud.Amount.describe()
```

```python
df.groupby('Class').mean()
```

```python
# Build a sample dataset containing similar distribution of normal transactions and Fraudulent Transactions
legit_New = legit.sample(n = 492)
```

```python
legit_New
```

```python
New_data = pd.concat([legit_New , fraud] , axis = 0 )
New_data.head()
```

```python
X = New_data.drop(columns = "Class" , axis = 1)
Y = New_data['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
clf=LogisticRegression(random_state=0)
clf.fit(X_train, Y_train)
```

```python
clf.score(X_train, Y_train)
```

```python
clf.score(X_test, Y_test)
```
